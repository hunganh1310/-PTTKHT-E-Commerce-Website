```python
# đường thẳng

import numpy as np
import matplotlib.pyplot as plt

# Dữ liệu giả lập: diện tích (m2) và giá nhà (triệu VND)
x = np.array([30, 40, 50, 60, 70, 80, 90, 100, 110, 120])
y = np.array([145, 213, 240, 320, 340, 420, 470, 510, 560, 600])

# Khởi tạo tham số
w = 0.0  # hệ số
b = 0.0  # bias
alpha = 0.0001  # learning rate
epochs = 2000   # số vòng lặp

n = len(x)
loss_history = []

# Gradient Descent
for epoch in range(epochs):
    y_pred = w * x + b
    error = y_pred - y
    loss = (1/n) * np.sum(error ** 2)
    loss_history.append(loss)

    # Tính đạo hàm
    dw = (2/n) * np.sum(error * x)
    db = (2/n) * np.sum(error)

    # Cập nhật tham số
    w -= alpha * dw
    b -= alpha * db

    # In ra mỗi 200 vòng lặp
    if epoch % 200 == 0:
        print(f"Epoch {epoch}: Loss = {loss:.2f}, w = {w:.4f}, b = {b:.4f}")

print("\nKết quả cuối cùng:")
print(f"w = {w:.4f}, b = {b:.4f}")

# Dự đoán với mô hình
y_pred = w * x + b

# Vẽ biểu đồ dự đoán so với dữ liệu gốc
plt.figure(figsize=(10,5))
plt.scatter(x, y, color='blue', label='Giá thật')
plt.plot(x, y_pred, color='red', label='Dự đoán')
plt.xlabel("Diện tích (m2)")
plt.ylabel("Giá nhà (triệu VND)")
plt.legend()
plt.title("Hồi quy tuyến tính bằng Gradient Descent")
plt.grid(True)
plt.show()

# Vẽ biểu đồ Loss
plt.plot(loss_history)
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Biểu đồ Loss theo thời gian")
plt.grid(True)
plt.show()
```

```python
# Hồi quy tuyến tính cho nhiều đặc trưng
import numpy as np

class LinearRegression:

    # Constructor
    def __init__(self, learning_rate, epochs, verbove):
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.verbove = verbove
        self.w = None
        self.b = None

    # hàm dự đoán predict:
    def predict(self, X):
        #y^ = w.x+b
        # X: m x n | w: n x 1
        return np.dot(X, self.w) + self.b

    def compute_loss(self, y_pred, y_true):
        return np.mean((y_pred - y_true)**2)

    def fit(self, X, y):

        # Khởi tạo trọng số ban đầu:
        m, n = X.shape
        self.w = np.zeros(n)
        self.b = 0


        # huấn luyện
        for epoch in range(self.epochs):

            # Tính y dự đoán
            y_pred = self.predict(X)

            # Tính sai số
            error =  y_pred - y

            #Tính gradient:
            dw = (2/m)*np.dot(X.T, error)
            db = (2/m)*np.sum(error)

            # Cập nhật trọng số:
            self.w = self.w - self.learning_rate * dw
            self.b = self.b - self.learning_rate * db

            loss = self.compute_loss(y_pred, y)

            #In ra màn hình
            if self.verbove and epoch % 20:
                print(f" Tập huấn luyện số {epoch}, Loss tương ứng là: {loss} ")


    def get_params(self):
            return self.w, self.b

X = np.array([
    [50, 2],
    [80, 3],
    [120, 4],
    [100, 3],
    [60, 2]
], dtype=np.float32)

y = np.array([100, 150, 250, 200, 120], dtype=np.float32)

model = LinearRegression(0.00001, 100, True)
model.fit(X, y)

w, b = model.get_params()
print("Trọng số:", w)
print("Bias:", b)




```

```python
import numpy as np

class LinearRegression:
    def __init__(self, learning_rate=0.01, epochs=1000, verbose=False):
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.verbose = verbose
        self.w = None
        self.b = None

    def predict(self, X):
        return np.dot(X, self.w) + self.b

    def compute_loss(self, y_true, y_pred):
        return np.mean((y_true - y_pred) ** 2)

    def fit(self, X, y):
        m, n = X.shape
        self.w = np.zeros(n)
        self.b = 0

        for epoch in range(self.epochs):
            y_pred = self.predict(X)
            error = y_pred - y

            # Tính gradient
            dw = (2 / m) * np.dot(X.T, error)
            db = (2 / m) * np.sum(error)

            # Cập nhật
            self.w -= self.learning_rate * dw
            self.b -= self.learning_rate * db

            if self.verbose and epoch % 100 == 0:
                loss = self.compute_loss(y, y_pred)
                print(f"Epoch {epoch}, Loss: {loss:.4f}")

    def get_params(self):
        return self.w, self.b
X = np.array([
    [50, 2],
    [80, 3],
    [120, 4],
    [100, 3],
    [60, 2]
], dtype=np.float32)

y = np.array([100, 150, 250, 200, 120], dtype=np.float32)

model = LinearRegression(learning_rate=0.0001, epochs=2000, verbose=True)
model.fit(X, y)

w, b = model.get_params()
print("Trọng số:", w)
print("Bias:", b)

```
